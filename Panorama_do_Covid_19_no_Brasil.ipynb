##Dicionário de Variáveis

**iso_code** - 3 letras representando o País, de acordo com o código ISO 3166-1 alpha-3;

**continent** - continente referente à localização geográfica;

**location** - localização geográfica;

**date** - data da observação;

**total_cases** - total de casos confirmados de COVID-19;

**new_cases** - novos casos confirmados de COVID-19;

**total_deaths** - total de mortes atribuídas ao COVID-19;

**new_deaths** - novas mortes atribuídas ao COVID-19;

**total_cases_per_million** - total de casos confirmados de COVID-19 por 1.000.000 de pessoas;

**new_cases_per_million** - novos casos confirmados de COVID-19 por 1.000.000 de pessoas;

**total_deaths_per_million** - total de mortes atribuídas ao COVID-19 por 1.000.000 de pessoas;

**new_deaths_per_million** - novas mortes atribuídas ao COVID-19 por 1.000.000 de pessoas;

**total_tests** - total de testes para COVID-19;

**new_tests** - novos testes para COVID-19;

**total_tests_per_thousand** - total de testes de COVID-19 por 1.000 pessoas;

**new_tests_per_thousand** - novos testes de COVID-19 por 1.000 pessoas;

**new_tests_smoothed** - dados suavizados de novos testes para COVID-19 (quando necessário em países que não relatam diariamente a quantidade de testes realizados);

**new_tests_smoothed_per_thousand** - dados suavizados de novos testes de COVID-19 por 1.000 pessoas;

**tests_units** - unidades utilizadas para relatar os dados de teste, em determinado local;

**stringency_index** - Índice de Estresse da Resposta do Governo. Valor dimensionado entre 0 e 100, que abrange 9 indicadores para contenção como fechamento de escolas e proibição de viagens;

**population** - população em 2020;

**population_density** - número de pessoas por Km2, no ano mais recente disponível;

**median_age** - idade média da população segundo projeção da ONU para 2020;

**aged_65_older** - Proporção da população com 65 anos ou mais, ano mais recente disponível;

**aged_70_older** - Percentagem da população com 70 anos ou mais em 2015;

**gdp_per_capita** - Produto interno bruto por pessoa; Equivalente ao poder de compra (dólares internacionais constantes de 2011) no último ano disponível;

**extreme_poverty** - Percentagem da população que vive em extrema pobreza, último ano disponível desde 2010;

**cvd_death_rate** - Taxa de mortalidade por doenças cardiovasculares em 2017;

**diabetes_prevalence** - Prevalência de diabetes (% da população entre 20 e 79 anos) em 2017;

**female_smokers** - Percentagem de mulheres que fumam, no último ano disponível;

**male_smokers** - Proporção de homens que fumam, no último ano disponível;

**handwashing_facilities** - Percentagem da população com instalações básicas de lavagem das mãos nas instalações, último ano disponível;

**hospital_beds_per_thousand** - Leitos de hospital por 1.000 pessoas, último ano disponível desde 2010;

**life_expectancy** - Expectativa de vida no nascimento em 2019;

## Países com mais casos

Vamos investigar os países com maior quantidade de casos totais registrados e maior registro de mortes até o dia desta análise (20/10/2020).

```
# Total de casos nesse último dia
df.loc[df.date=='2020-10-20'].sort_values(by='total_cases', ascending=False)
```

Vemos acima que no dia 20 de Outubro de 2020, os países líderes com mais casos de Covid-19 são:

1.   Estados Unidos - 8.214.755 casos;
2.   India - 7.597.063 casos;
3.   Brasil - 5.250.727 casos.

Vamos dá uma olhada nos casos de mortes.


```
# Total de mortos
df.loc[df.date=='2020-10-20', ['location','total_deaths']].sort_values(by='total_deaths', ascending=False)
```

Em relação às mortes, os países que lideravam aos 20 de Outubro de 2020 eram:

* Estados Unidos - 220.133 mortes;
* Brasil - 154.176 mortes;
* India - 115.197  mortes.

Para melhorar a visualização, vamos plotar um gráfico de barras.

### Os 5 países com mais mortes

```
# Gráfico de barras para o total de mortes

top5_totalMortes = df.loc[df.date=='2020-10-20',['location','total_deaths']].sort_values(by='total_deaths', ascending=False)[1:6]
fig, ax = plt.subplots(figsize=(10,5))
ax.bar('location','total_deaths',data=top5_totalMortes)
ax.set_title('Os 5 países com mais mortes')
plt.tight_layout()
```

### A progressão do vírus

A maior preocupação das organizações de saúde é a progressão do número de casos e mortes no mundo.

Uma das melhores formas de visualizarmos a evolução de um dado ao longo do tempo, é o gráfico de linhas. Vamos plotar esse gráfico para demonstrar a evolução de mortes no mundo.



```
# gráfico de linha para mortes no mundo
df.loc[df.location=='World','total_deaths'].reset_index(drop=True).plot(figsize=(10,7));
```

## Explorando os dados do Brasil

Agora que já vimos as imformações do mundo, vamos olhar os acontecimentos no Brasil.


```
# Gerando um dataframe só com os dados do Brasil
df_Brasil = df.loc[df.location=='Brazil']
df_Brasil.head()
```

### Os primeiros registros no Brasil

```
print('Data do primeiro registro: ',df_Brasil.date.min())
print('Data do último registro: ',df_Brasil.date.max())
```

Acima vemos que no dataset do Brasil inclui dados desde 31-12-2019 até 20-10-2020.

Vamos verificar quando foi registrada o primeiro caso e a primeira morte no Brasil.

### Primeiro caso e primeira morte

```
# Identificando quando foi registrado o primeiro caso no Brasil
df_Brasil.loc[df_Brasil['new_cases'] != 0.0].head(1)
```

```
# Identificando quando foi registrado a primeira morte no Brasil
df_Brasil.loc[df_Brasil['new_deaths'] != 0.0].head(1)
```

```
print(f'O primeiro caso foi registrado em: {df_Brasil.date.loc[6638]}')
print(f'A primeira morte foi registrada em: {df_Brasil.date.loc[6659]}')
```

### Dias entre o primeiro caso e a primeira morte

Vamos ver quanto tempo levou do aparecimento do primeiro caso até a primeira morte.

```
# Calculando quantos dias do primeiro caso até a primeira morte
from datetime import date
dias = df_Brasil.date.loc[6659] - df_Brasil.date.loc[6638]
print(f'Se passaram {dias.days} dias entre o primeiro caso e a primeira morte no Brasil')
```


### Gráfico comparativo de casos e mortes

Para finalizar, vamos plotar um gráfico de linha para compararmos a evolução dos casos e mortes no Brasil.

```
x = list(range(295))
z = df_Brasil['total_cases']
z2 = df_Brasil['total_deaths']

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 5))
ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, z)

ax2.set_title('Mortes no Brasil')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Mortes')
ax2.plot(x, z2)

plt.tight_layout()
```
